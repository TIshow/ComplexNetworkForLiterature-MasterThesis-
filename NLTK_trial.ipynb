```python
fdist = nltk.FreqDist(w.lower() for w in text)
```

```python
fdist.keys()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20,10))
fdist.plot(50, cumulative=True)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
symbols = ["'", '"', '`', '.', ',', '-', '!', '?', ':', ';', '(', ')', '...', "'s"]
```

```python
fdist = nltk.FreqDist(w.lower() for w in text if w.lower() not in stopwords + symbols)

```

```python
plt.figure(figsize=(10,10))
fdist.plot(50, cumulative=True)
```

```python
#登場人物の名前である'mark'，'eduardo'，'sean'が最も頻出しており，
#'room'や'night'といった単語の出現頻度も高いことが分かる． 
#一部"'s"や"n't"といったアポストロフィー付きの短いトークンも混ざっているが，
#これらは"Mark's"や"didn't"のような単語が分割された結果出てきたトークンである．
#本来はこれらも取り除いてから頻度分布を調べた方がいいだろう．
```
