```python
#Suspense and Thriller
import nltk
from nltk.corpus import stopwords

raw1 = open('The_Adventure_of_the_Illustrious_Client.txt').read()
```

```python
tokens1 = nltk.word_tokenize(raw1)
sentence1 = nltk.sent_tokenize(raw1)
```

```python
tokens1
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer() #to make base form

tokens_low = [w.lower() for w in tokens1]

#The number of sort of words
len(set(tokens_low))

#複数形や過去形を別の単語としてカウントしてしまっているため，実際の異なり語数はさらに少ない
```

```python
tokens_low
```

```python
list = [tokens1]

text = []
for i in list:
    text.append(nltk.Text(i))
    
text
```

```python
f_list = []
for i in text:
    f_list.append(nltk.FreqDist(w.lower() for w in i))

stopwords = nltk.corpus.stopwords.words('english')
symbols = ["'", '"', '`', '.', ',', '-', '!', '?', ':', ';', '(', ')', '...', "'s", '’', '”', '“','–','—', '…', '*' ,'``', "''",'--','mr.',"n't", 'dr.' ]
```

```python
#Eliminate stopwords and symbols

import matplotlib.pyplot as plt
%matplotlib inline

f_list_new = []
for i in text:
    f_list_new.append(nltk.FreqDist(w.lower() for w in i if w.lower() not in stopwords + symbols))

for i in f_list_new:
    plt.figure(figsize=(10,5))
    i.plot(50)
```

```python
#順位チェック
sorted(f_list_new[0].items(), key=lambda x: x[1], reverse=True)
```

```python
f_list_new
```

```python
tokens_low_new = []
for i in tokens_low:
    if i not in stopwords + symbols:
        tokens_low_new.append(i)
```

```python
tokens_low_new
```

```python
indexes = []
for i in tokens_low_new:
    for n,x in enumerate(tokens_low_new):
        if x == i:
            indexes.append(n)
            
    
```

```python
indexes
```

```python
a = sorted(f_list_new[0].items(), key=lambda x: x[1], reverse=True)
```

```python
new_a = []
for i in a:
    if i[1] > 1:
        new_a.append(i)
```

```python
new_a
```

```python
indexes = []
for i in new_a:
    indexes.append(i)
    for n,x in enumerate(tokens_low_new):
        if x == i[0]:
            indexes.append(n)
            
```

```python
holmes_indexes = []
for n,x in enumerate(tokens_low_new):
        if x == 'holmes':
            holmes_indexes.append(n)
```

```python
holmes_indexes
```

```python
holmes_distance = []
count = 0
for i in holmes_indexes:
    if i != holmes_indexes[-1]:
        distance = holmes_indexes[count+1] - i
        holmes_distance.append(distance)
        
        count +=1
    
```

```python
holmes_distance
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.plot(sorted(holmes_distance, reverse=True))

plt.ylabel('s value')
plt.xlabel('words (Descending order of s value)')
plt.show()
```
