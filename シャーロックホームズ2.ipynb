```python
list = [tokens1]

text = []
for i in list:
    text.append(nltk.Text(i))
    
text
```

```python
f_list = []
for i in text:
    f_list.append(nltk.FreqDist(w.lower() for w in i))

stopwords = nltk.corpus.stopwords.words('english')
symbols = ["'", '"', '`', '.', ',', '-', '!', '?', ':', ';', '(', ')', '...', "'s", '’', '”', '“','–','—', '…', '*' ,'``', "''",'--','mr.',"n't", 'dr.','sir' ]
```

```python
#Eliminate stopwords and symbols

import matplotlib.pyplot as plt
%matplotlib inline

f_list_new = []
for i in text:
    f_list_new.append(nltk.FreqDist(w.lower() for w in i if w.lower() not in stopwords + symbols))

for i in f_list_new:
    plt.figure(figsize=(20,10))
    i.plot(50)
```

```python
#いけそう。文ごとに取り出し、それぞれの単語を原型に戻す。それを再び再構築する！
for sentence in sentence_l1:
    form = [ps.stem(word)  for word in sentence.split() if not word in stopwords + symbols]
    print (form)
```

```python
#上の本番
form_sent = []
for sentence in sentence_l1:
    form = [ps.stem(word)  for word in sentence.split() if not word in stopwords + symbols]
    form_sent.append(" ".join(form))
```

```python
form_sent
```

```python
#ピリオドやカンマを消しまーす。Delte period and comma...めんどい...
form_sent2 = []

for i in form_sent:
    tokens = nltk.word_tokenize(i)
    form = [word for word in tokens if not word in stopwords + symbols]
    form_sent2.append(" ".join(form))
    
```

```python
form_sent2
```

```python
#次にこの新しく、原型に戻し、stopwordを取り除いた単語数をグラフ化したい
text = []

for i in form_sent2:
    tokens = nltk.word_tokenize(i)
    for j in tokens:
        text.append(j)
```

```python
#新しく、原型に戻し、stopwordを取り除いた単語数をグラフ化

import matplotlib.pyplot as plt
%matplotlib inline

f_list_new = []
for i in text:
    f_list_new.append(nltk.FreqDist(w.lower() for w in i if w.lower() not in stopwords + symbols))

for i in f_list_new:
    plt.figure(figsize=(20,10))
    i.plot(50)
```

```python
# let's make new list only with words whose frequency  > 1

moreThan2_freq = []

for i,j in f_list_new[0].items():
    if j > 1:
        moreThan2_freq.append(i)
```

```python
moreThan2_freq
```

```python
# 頻出単語top2のgameshとsaidが同時に現れる文を、ピックアップしカウントするプログラムを全ての単語に割り当てる試み。

count_s = {}

for wn1 in range(0, len(moreThan2_freq)-1):
    for wn2 in range(wn1+1, len(moreThan2_freq)):
        s = 0
    
        for sent in form_sent2:
            index1 = 0
            index2 = 0
            
            #remove duplicate
            if moreThan2_freq[wn1] != moreThan2_freq[wn2]:
                
                if moreThan2_freq[wn1] in sent.split():
                    index1 += 1
                if moreThan2_freq[wn2] in sent.split():
                    index2 += 1
              
                if index1 > 0 and index2 > 0:
                    s += 1
                
        count_s.update({( moreThan2_freq[wn1], moreThan2_freq[wn2]) :s })
        
```

```python
# could
indexes = []
for i in new_a:
    indexes.append(i)
    for n,x in enumerate(tokens_low_new):
        if x == i[0]:
            indexes.append(n)
            
could_indexes = []
for n,x in enumerate(tokens_low_new):
        if x == 'could':
            could_indexes.append(n)
            
could_distance = []
count = 0
for i in could_indexes:
    if i != could_indexes[-1]:
        distance = could_indexes[count+1] - i
        could_distance.append(distance)
        
        count +=1


plt.figure(figsize=(20,10))
plt.plot(sorted(could_distance, reverse=True))

plt.ylabel('s value')
plt.xlabel('words (Descending order of s value)')
plt.show()
```
