```python
#Suspense and Thriller
import nltk
from nltk.corpus import stopwords

raw1 = open('The_Adventure_of_the_Illustrious_Client.txt').read()
```

```python
tokens1 = nltk.word_tokenize(raw1)
sentence1 = nltk.sent_tokenize(raw1)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer() #to make base form

tokens_low = [w.lower() for w in tokens1]

#The number of sort of words
len(set(tokens_low))

#複数形や過去形を別の単語としてカウントしてしまっているため，実際の異なり語数はさらに少ない
```

```python
tokens_low
```

```python
print (len(set(tokens_low))/len(tokens1))
```

```python
list = [tokens1]

text = []
for i in list:
    text.append(nltk.Text(i))
    
text
```

```python
f_list = []
for i in text:
    f_list.append(nltk.FreqDist(w.lower() for w in i))

stopwords = nltk.corpus.stopwords.words('english')
symbols = ["'", '"', '`', '.', ',', '-', '!', '?', ':', ';', '(', ')', '...', "'s", '’', '”', '“','–','—', '…', '*' ,'``', "''",'--','mr.',"n't", 'dr.' ]
```

```python
f_list
```

```python
#Eliminate stopwords and symbols

import matplotlib.pyplot as plt
%matplotlib inline

f_list_new = []
for i in text:
    f_list_new.append(nltk.FreqDist(w.lower() for w in i if w.lower() not in stopwords + symbols))

for i in f_list_new:
    plt.figure(figsize=(20,10))
    i.plot(50)
```

```python
sorted(f_list_new)
```

```python
#順位チェック
sorted(f_list_new[0].items(), key=lambda x: x[1], reverse=True)
```

```python
#全部小文字にした時の単語の羅列をもう一度見ておく。
tokens_low
```

```python
#Turn all words to basic form
bf = [ps.stem(word) for word in tokens_low if not word in stopwords + symbols]
```

```python
bf
```

```python
#Textを使うためにreviewをlistにぶち込む
bf = [bf]
```

```python
#原型に戻した単語で、再び文を再構築する。
text = []
for i in bf:
    text.append(nltk.Text(i))
    
text
```

```python
#Textからwords listを作る。
for i in text:
    words = [word for word in i]
    
words
```

```python
# Sentence bu sentence
sentence1
```

```python
#文章を小文字化
sentence_l1 = [w.lower() for w in sentence1]
sentence_l1
```

```python
#いけそう。文ごとに取り出し、それぞれの単語を原型に戻す。それを再び再構築する！
for sentence in sentence_l1:
    form = [ps.stem(word)  for word in sentence.split() if not word in stopwords + symbols]
    print (form)
```

```python
form
```

```python
#上の本番
form_sent = []
for sentence in sentence_l1:
    form = [ps.stem(word)  for word in sentence.split() if not word in stopwords + symbols]
    form_sent.append(" ".join(form))
```

```python
form_sent
```

```python
#ピリオドやカンマを消しまーす。Delte period and comma...めんどい...
form_sent2 = []

for i in form_sent:
    tokens = nltk.word_tokenize(i)
    form = [word for word in tokens if not word in stopwords + symbols]
    form_sent2.append(" ".join(form))
    
```

```python
form_sent2
```

```python
#次にこの新しく、原型に戻し、stopwordを取り除いた単語数をグラフ化したい
text = []

for i in form_sent2:
    tokens = nltk.word_tokenize(i)
    for j in tokens:
        text.append(j)
```

```python
text
```

```python
text = [text]
```

```python
#新しく、原型に戻し、stopwordを取り除いた単語数をグラフ化

import matplotlib.pyplot as plt
%matplotlib inline

f_list_new = []
for i in text:
    f_list_new.append(nltk.FreqDist(w.lower() for w in i if w.lower() not in stopwords + symbols))

for i in f_list_new:
    plt.figure(figsize=(20,10))
    i.plot(50)
```

```python
f_list_new
```

```python
#順位チェック
sorted(f_list_new[0].items(), key=lambda x: x[1], reverse=True)
```

```python
# 頻出単語top2のgameshとsaidが同時に現れる文を、ピックアップしカウントするプログラムを全ての単語に割り当てる試み。

count_s = {}

for wn1 in range(0, len(text[0])-1):
    for wn2 in range(wn1+1, len(text[0])):
        s = 0
    
        for sent in form_sent2:
            index1 = 0
            index2 = 0
            
            #remove duplicate
            if text[0][wn1] != text[0][wn2]:
                
                if text[0][wn1] in sent.split():
                    index1 += 1
                if text[0][wn2] in sent.split():
                    index2 += 1
              
                if index1 > 0 and index2 > 0:
                    s += 1
                
        count_s.update({( text[0][wn1], text[0][wn2]) :s })
        
```

```python
count_s
```

```python
#順位チェック
sorted(count_s.items(), key=lambda x: x[1], reverse=True)
```

```python
#namely

new = {tuple(sorted(x)):y for x, y in count_s.items()}
```

```python
#順位チェック again !
new2 = dict(sorted(new.items(), key=lambda x: x[1], reverse=True))
new2
```

```python
new3 = dict((k, v) for k, v in new2.items() if v >= 2)
new3
```

```python
#順位チェック again !
new4 = dict(sorted(new3.items(), key=lambda x: x[1], reverse=True))
new4
```
